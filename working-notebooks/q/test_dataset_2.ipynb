```python
import pandas as pd
```

```python
reviews = pd.read_csv("/mnt/data/public/bgg/bgg-15m-reviews.csv")
df = pd.read_csv('/mnt/data/public/bgg/games_detailed_info.csv')
test1 = pd.read_csv("/mnt/data/public/bgg/2022-01-08.csv")
```

```python
test1.rename(columns={'Name':'name'}, inplace=True)
test1.rename(columns={'Users rated':'number_of_reviews'}, inplace=True)
df.rename(columns={'primary':'name'}, inplace=True)
```

```python
new_df = test1.copy(deep=True)[['name']]
new_df = new_df.merge(df[['name','description']], on='name', how='left')
new_df = new_df.merge(test1[['name','Average','number_of_reviews']], on='name', how='left')
```

```python
new_df.dropna(inplace=True)
new_df.drop_duplicates(inplace=True)
```

```python
new_df.iloc[:10000, :].to_csv('../../data/boardgames_final.csv')
```
