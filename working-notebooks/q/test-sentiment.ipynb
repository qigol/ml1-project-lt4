```python
import pandas as pd
```

```python
reviews = pd.read_csv("/mnt/data/public/bgg/bgg-15m-reviews.csv")
```

```python
reviews.head()
```

```python
def get_average_review(game):
    return reviews[reviews['name']==game]['rating'].mean()
```

```python
get_average_review('Pandemic', reviews)
```

```python
review_df = pd.DataFrame(data={'name': reviews['name'].unique()})
```

```python
review_df['average_review'] = review_df['name'].apply(get_average_review)
```
